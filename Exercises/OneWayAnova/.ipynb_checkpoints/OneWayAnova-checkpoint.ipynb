## Analysis of Variance

#### Analysis of variance is a collection of statistical models and their associated estimation procedures used to analyze the differences among group means in a sample.

Is there a difference in mathematics scores of 3 sections of class 19 in Nirmal Vidhyalaya ? A random sample of scores are taken from 3 different sections

```python
sec1 = [17,16,11,15,10,14,13]
sec2 = [12,19,19,17,13,20,18]
sec3 = [10,12,17,9,7,9,11]
```

```python
scores = dict(sec1=sec1,sec2=sec2,sec3=sec3)
scores
```

```python
# Calculating the F statistic directly using the f one way anova

from scipy import stats
stats.f_oneway(sec1,sec2,sec3)
```

* Null Hypothesis H0: Mu1 = Mu2 = Mu3
* Alternate H1 : all not equal
* Since p < 0.5 Reject H0
* Also F<sub>statistic</sub> > F<sub>Critical</sub> See the description Below

```python
# TO calculate the f statistic from dfn and dfd alpha = .05
# For one-way ANOVA, the ratio of the between-group variability to the within-group variability
# follows an F-distribution when the null hypothesis is true.
# For our example 
a=3
n=7
stats.f.ppf(1-.05,dfn=a-1,dfd=a*(n-1))
```

```python
from IPython.display import Image
Image('F-Distribution.JPG')
```

```python
# Better Pythonic way of doing it.
import pandas as pd
import numpy as np
import math, scipy.stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
scores_data = pd.DataFrame(scores)
```

```python
scores_data
```

```python
# converting the data into more realistic format 
realscores = pd.melt(scores_data.reset_index(), value_vars=['sec1','sec2','sec3'])
realscores.columns=['Groups','Observations']
```

```python
realscores.head()
```

```python
model = ols('Observations ~ Groups', data=realscores).fit()
anova_table = sm.stats.anova_lm(model,type=1)
```

```python
anova_table
```

```python
########################## AFTER ANOVA, IF NULL HYPOTHESIS IS REJECTED ##############################
# WE CAN PERFORM THE POST HOC TEST, LSD TEST OR TUKEY's TEST TO DO A PAIRWISE ANALYSIS WHICH CheCKS 
# WHICH GROUP MEAN IS SIGNIFICANTLY DIFFERENT FROM WHICH MEAN 
```

## Post Hoc Analysis 

### Fischer's Least significant distance  LSD mthod

```python
# MANUAL CALCULATION 
# CONSIDERING SAME DATA AS EXAMPLE 

def LSDcalc(a,n,MSE,alpha=0.025):
    t = -1 * stats.t.ppf(0.025,a*(n-1))
    return(t + math.sqrt(2*MSE/n))

Image('LSDFormula.JPG',width=300,height=300)
```

```python
a = len(np.unique(realscores['Groups']))
LSD = LSDcalc(a,n=7,MSE=8.873016)
LSD
```

```python
# NOW THAT LSD is 3.69 We have to check all pairs absolute means and compare with LSD
means = realscores.groupby(['Groups']).mean()
means
```

```python
## Generate Pairwise Comparison
```

```python
#| sec 1 - sec 2 | = 13.714286 - 16.857143
#| sec 1 = sec 3 | = 13.714286 - 10.714286
#| sec 2 - sec 3 | = 16.857143 - 10.714286

oneandtwo = abs(13.714286 - 16.857143)
oneandthree = abs(13.714286 - 10.714286)
twoandthree = abs(16.857143 - 10.714286)
```

```python
oneandtwo, oneandthree, twoandthree
```

##### Pair twoandthree is greater than LSD Therefore mu2 NOT EQUAL TO mu3 

### Tukey Kramer Critical Range test

```python
# Determines the significant distance between the means.
Image('TukeysCompare.JPG',width=500)
```

```python
# Tukey gives the formula to calcualate the critical range, and if the absolute difference of means of pairs is
# greater than critical range, the means of the pairs are significantly difference
Image('TukeysCriticalRange.JPG',width=400)
```

```python
# Qu is the value of studentized range distribution with c and n-c degrees of freedon for desired Alpha
# MSW = MSE = Mean Square Within
# c= number of columns / Levels / Groups
# n = total number of sample size. 
```

####  Manual Calculation

```python
# Qu is calculated from the Qu Table 
# c = 3 
# n-c = 21-3 = 18
Image('Qtable.JPG')
```

```python
Qu = 3.61 # for Alpha = 0.05  4.7 for 0.01 Alpha level of significance 
```

```python
# Critical Range 
MSE=8.873016
CriticalRange = Qu * math.sqrt((MSE/2)*( (1/7) + (1/7) ) )
CriticalRange
```

```python
# Again 
oneandtwo = abs(13.714286 - 16.857143)
oneandthree = abs(13.714286 - 10.714286)
twoandthree = abs(16.857143 - 10.714286)
oneandtwo, oneandthree, twoandthree
```

```python
# Only Mu2 and Mu3 have absolute mean difference > 4.06 
# Therefore Mu2 and Mu3 are different 
```

####  Using Python Library

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison
MC = MultiComparison(realscores['Observations'],realscores['Groups'])
MCRESULT = MC.tukeyhsd(0.05)
MCRESULT.summary()
```

```python
#### Sec2 and Sec3 group is rejected
#### Therefore Sec2 NOT equal to Sec3 
```
